```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
# 모델 로드
model8 = YOLO('yolov8n.pt')
model8o = YOLO('yolov8n-obb.pt')
model11 = YOLO('yolo11n.pt')
model11o = YOLO('yolo11n-obb.pt')

# 테스트할 이미지 경로 설정
img_path = './data/images/led.jpg'
```

```python
# 예측 실행
results_8 = model8.predict(img_path)
results_8o = model8o.predict(img_path)
results_11 = model11.predict(img_path)
results_11o = model11o.predict(img_path)
```

```python
# 결과 시각화
img8 = results_8[0].plot()
img_rgb8 = cv2.cvtColor(img8, cv2.COLOR_BGR2RGB)

img8o = results_8o[0].plot()
img_rgb8o = cv2.cvtColor(img8o, cv2.COLOR_BGR2RGB)

img11 = results_11[0].plot()
img_rgb11 = cv2.cvtColor(img11, cv2.COLOR_BGR2RGB)

img11o = results_11o[0].plot()
img_rgb11o = cv2.cvtColor(img11o, cv2.COLOR_BGR2RGB)
```

```python
# 결과 이미지 출력
fig, ax = plt.subplots(2, 2)

ax[0][0].imshow(img_rgb8)
ax[0][0].set_title('YOLOv8 Detection')
ax[0][0].axis('off')

ax[0][1].imshow(img_rgb8o)
ax[0][1].set_title('YOLOv8 OBB')
ax[0][1].axis('off')

ax[1][0].imshow(img_rgb11)
ax[1][0].set_title('YOLOv11 Detection')
ax[1][0].axis('off')

ax[1][1].imshow(img_rgb11o)
ax[1][1].set_title('YOLOv11 OBB')
ax[1][1].axis('off')

plt.tight_layout()
plt.show()
```
